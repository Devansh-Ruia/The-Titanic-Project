```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
#trains.csv file contains passanger details
data=pd.read_csv('/kaggle/input/titanic-competition-datasets/titanic/train.csv')
```

```python
data.count()
```

```python
#code to drop empty or abnormal columns
data=data.drop(['Name','Ticket', 'Cabin', 'Embarked'],axis=1)
#Filling missing values
data['Age'].fillna(data['Age'].median(),inplace=True)
#Converting catogory data to numerical data
label_encoder=LabelEncoder()
data['Sex']=label_encoder.fit_transform(data['Sex'])
```

```python
x=data.drop(['Survived'],axis=1)
y=data['Survived']
```

```python
#Training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test
```

```python
#Instance for deciusion tree clasifier
model=DecisionTreeClassifier()
#Fit the data into test
model.fit(x_train,y_train)
```

```python
#Prediction on a test value
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
output=pd.DataFrame({'PassengerId':x_test.PassengerId,'Survived':y_pred})
output.to_csv('Titanic_project1.csv',index=False)
```
